## **2.Feature Engeneering**

### 2.1 Variable Objetivo **GA**: Goles + Asistencias

Generamos nuestra variable objetivo y la tranformamos a numero entero para mejorar la prediccion del modelo.

```python
# Se crea la variable a analizar: Goles + Asistencias c/90 min
df_pl['GA'] = df_pl['Goals'] + df_pl['Assists']

# Redondea la columna 'GA' a cero decimales
df_pl['GA'] = df_pl['GA'].round(0)

# Convierte la columna 'GA' a tipo de datos entero
df_pl['GA'] = df_pl['GA'].astype(int)

# Muestra las primeras filas del DataFrame para verificar
df_pl.head()
```

### 2.2 Recategorizamos Posiciones

```python
# Creamos una función para mapear las posiciones de forma mas general
def map_position_to_category(position):
    if "FW" in position or "FWMF" in position or "FWDF" in position:
        return "Delanteros"
    elif "DF" in position or "DFFW" in position or "DFMF" in position:
        return "Defensores"
    elif "MF" in position or "MFFW" in position or "MFDF" in position:
        return "Mediocampistas"
    elif "GK" in position or "GKMF" in position:
        return "Arqueros"
    else:
        return "Otra"

# Aplicamos la función para crear la nueva columna "Posicion"
df_pl['Posicion'] = df_pl['Pos'].apply(map_position_to_category)

# Eliminamos Pos
columnas = ['Pos']
df_pl = df_pl.drop(columns=columnas)
```

### 2.3 Recategorizamos Edades

*Recordar ver de eliminar variable Edad[texto del enlace

```python
# Definir una función para asignar las categorías 'Young', 'Mid' y 'Old' en base a la edad
def categorizar_edad(edad):
    if edad <= 22:
        return 'Young'
    elif 23 <= edad <= 30:
        return 'Mid'
    else:
        return 'Old'

# Crear la columna 'Old' utilizando la función y apply
df_pl['Old'] = df_pl['Age'].apply(categorizar_edad)
```

### 2.4 Normalizacion de Features



```python
from sklearn.preprocessing import StandardScaler

# Crea un objeto StandardScaler
scaler = StandardScaler()

# Ajusta y transforma la variable LgRk
df_pl['sq_Lg_Rk_norm'] = scaler.fit_transform(df_pl[['sq_Lg_Rk']].values)

# Ajusta y transforma la variable LgRk
df_pl['sq_GF_norm'] = scaler.fit_transform(df_pl[['sq_GF']].values)

# Ajusta y transforma la variable sq_Pts
df_pl['sq_Pts_norm'] = scaler.fit_transform(df_pl[['sq_Pts']].values)

# Lista de columnas a eliminar
columnas_a_eliminar = ['sq_Lg_Rk', 'sq_GF', 'sq_Pts']

# Elimina las columnas del DataFrame
df_pl = df_pl.drop(columns=columnas_a_eliminar)
```

### 5.3 Calculo de Metricas

```python
#Definir una función para evaluar rendimiento de modelos

def evaluate_model(y_true, y_pred, model_name):
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)

    # Redondear los resultados a dos decimales
    rmse = round(rmse, 2)
    r2 = round(r2, 2)
    mae = round(mae, 2)

    print(f'{model_name}: RMSE={rmse}, R2={r2}, MAE={mae}')
```

```python
#Evaluación de modelos en la función que definimos

evaluate_model(y_test, y_pred_linear, 'Regresión Lineal')

evaluate_model(y_test, y_pred_rf, 'Random Forest')

evaluate_model(y_test, y_pred_gb, 'Gradient Boosting')
```

**Tenemos overfitting?**

```python
#Definir función para evaluar métricas de entrenamiento vs métricas de prueba

def evaluate_model_train_test(model, X_train, y_train, X_test, y_test, model_name): #Nombre de función

    # Predicciones en conjuntos de entrenamiento y prueba

    y_train_pred = model.predict(X_train) #predicción de entrenamiento

    y_test_pred = model.predict(X_test) #predicción de prueba



    # Métricas en conjuntos de entrenamiento

    rmse_train = mean_squared_error(y_train, y_train_pred, squared=False)

    r2_train = r2_score(y_train, y_train_pred)

    mae_train = mean_absolute_error(y_train, y_train_pred)



    # Métricas en conjuntos de prueba

    rmse_test = mean_squared_error(y_test, y_test_pred, squared=False)

    r2_test = r2_score(y_test, y_test_pred)

    mae_test = mean_absolute_error(y_test, y_test_pred)

    # Crear un DataFrame con los resultados

    results_df = pd.DataFrame({

        'Modelo': [model_name, model_name],

        'Conjunto': ['Entrenamiento', 'Prueba'],

        'RMSE': [rmse_train, rmse_test],

        'R2': [r2_train, r2_test],

        'MAE': [mae_train, mae_test]

    })



    return results_df
```

```python
#Ingresar valores a la función de métricas entrenamiento vs prueba

random_forest_results = evaluate_model_train_test(random_forest, X_train, y_train, X_test, y_test, 'Random Forest')

gradient_boosting_results = evaluate_model_train_test(gradient_boosting, X_train, y_train, X_test, y_test, 'Gradient Boosting')



# Concatenar los resultados en un único DataFrame

all_results = pd.concat([ random_forest_results, gradient_boosting_results], ignore_index=True)



# Imprimir el DataFrame de resultados

print(all_results)
```

### 5.5 Optimizacion de Parametros

```python
from xgboost import XGBRegressor
from sklearn.experimental import enable_halving_search_cv  # noqa
from sklearn.model_selection import HalvingGridSearchCV
import warnings
from sklearn.exceptions import UndefinedMetricWarning

# Ignorar ciertas advertencias
warnings.filterwarnings("ignore", category=UserWarning, message="The least populated class in y has only 1 members")
warnings.filterwarnings("ignore", category=UndefinedMetricWarning, message="R^2 score is not well-defined with less than two samples.")

# Definir el modelo
modelo_xgb = XGBRegressor()

# Definir los parámetros a optimizar
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

# Definir la estrategia de validación cruzada
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Inicializar el objeto HalvingGridSearchCV
halving_search = HalvingGridSearchCV(
    modelo_xgb, param_grid, scoring='r2', cv=skf, factor=2, resource='n_samples',
    max_resources=300, aggressive_elimination=True, random_state=42
)

# Realizar la búsqueda de hiperparámetros
halving_search.fit(X, y)

# Imprimir los mejores parámetros encontrados
print("Mejores hiperparámetros:", halving_search.best_params_)
```
